```python
def summary(df):
    summy = pd.DataFrame(df.dtypes, columns=['data type'])
    summy['#missing'] = df.isnull().sum().values
    summy['Duplicate'] = df.duplicated().sum()
    summy['#unique'] = df.nunique().values
    return summy
```
